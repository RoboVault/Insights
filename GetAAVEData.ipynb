```python
from shroomdk import ShroomDK
import json
import pandas as pd 

sdk = ShroomDK("5c0c5e4b-47eb-4cc2-835d-a2ddad6d1a60")

Years = [2020,2021,2022]
Months = range(1,13)

table = "ethereum.aave.ez_deposits"

df_list = []

maxDate = 202302 

def getData(table, Years, Months):
    tableName = table.replace(".", "_")
    print("Grabbing Data For : " + tableName)
    for Year in Years : 
        for Month in Months : 
            if Year*100 + Month < maxDate : 

                print("Getting Data " + str(Year) + " - " + str(Month))
                sql = f"""
                    SELECT * 
                    FROM {table}
                    WHERE  
                      year(block_timestamp) = {Year} AND 
                      month(block_timestamp) = {Month} AND  
                      AAVE_VERSION = 'Aave V2'
                      
                """

                results = sdk.query(sql)
                data = json.loads(results.json())
                df = pd.DataFrame(data['records'])
                print("Total Records : " + str(len(df)))
                df_list.append(df)

        
    finalDf = pd.concat(df_list)    
    finalDf.to_csv(tableName + ".csv")
```

```python
#tables = ["ethereum.aave.ez_deposits", "ethereum.aave.ez_liquidations", "ethereum.aave.ez_repayments", "ethereum.aave.ez_borrows", "ethereum.aave.ez_withdraws"]

tables = ["ethereum.aave.ez_withdraws"]

for table in tables : 
    getData(table, Years, Months)
```
